## Region wise analysis

* **We will make separate dataframe for each region and analyse them**
* **Also we will define some helper functions which we will be using repeated in the further analysis**

```
## making a separate df for each region 
west_df = df[df['Region'] == 'West']
east_df = df[df['Region'] == 'East']
central_df = df[df['Region'] == 'Central']
south_df = df[df['Region'] == 'South']


dfs = [
       west_df,
       east_df,
       central_df,
       south_df
]
```

```
## Helper functions for analysis

def barplotter(region_df, xcol, ycol, palette):
    plt.figure(figsize=(10,5), dpi=100)
    plt.xticks(rotation = 90)
    sns.barplot(x = xcol,
                y = ycol,
                data=region_df,
                palette=palette)

def average(region_df, col1, col2):
    return region_df.groupby(col1)[col2].mean()

def total(region_df, col1, col2):
    return region_df.groupby(col1)[col2].sum()

def pie_plotter_total(title, region_df, label_col, value_col):
    plt.title(title)
    colors = ['lightseagreen', 'mediumpurple', 'aquamarine', 'lawngreen', 'mediumorchid']
    plt.pie(
            total(region_df, label_col, value_col),
            labels= total(region_df, label_col, value_col).index,
            wedgeprops=dict(width = 0.5),
            shadow=True,
            colors=colors,
            startangle= 90,
            autopct='%1.1f%%'
    )
     
```

### Ship Mode v/s Sales (Region-wise)

```
barplotter(west_df, 'Ship Mode', 'Sales', 'tab10');
```

```
barplotter(east_df, 'Ship Mode', 'Sales', 'hls');
```

```
barplotter(central_df, 'Ship Mode', 'Sales', 'Paired');
```

```
barplotter(south_df, 'Ship Mode', 'Sales', 'mako');
```

* **In West region, highest sales were made from Second class Ship mode**
* **In East region, highest sales were made from Same day Ship mode**
* **in Central and South region, highest sales were made from Second class Ship mode**

### Category v/s Sales (Region-wise)

```
plt.figure(figsize= (10,10), dpi = 100)

plt.subplot(2,2,1)
pie_plotter_total('West', west_df, 'Category', 'Sales')

plt.subplot(2,2,2)
pie_plotter_total('East', east_df, 'Category', 'Sales')

plt.subplot(2,2,3)
pie_plotter_total('Central', central_df, 'Category', 'Sales')

plt.subplot(2,2,4)
pie_plotter_total('South', south_df, 'Category', 'Sales')
```

**Highest total sales were made in Technology category in each region**

```
plt.figure(figsize=(10,5), dpi=100)
sns.barplot(x = 'Sales', 
            y = 'Region', 
            hue='Category', 
            data=df,
            palette='husl');
```

**HIghest average sales were also made in Technology category**

```
barplotter(west_df, 'Sub-Category', 'Sales', 'flare')
```

```
barplotter(east_df, 'Sub-Category', 'Sales', 'hls')
```

```
barplotter(central_df, 'Sub-Category', 'Sales', 'husl')
```

```
barplotter(south_df, 'Sub-Category', 'Sales', 'magma')
```

* **In every region except South, goods from Copiers sub-category have made the highest average sales**
* **in South region, goods from Machine sub-category have made the highest average sales**

### Segment v/s Profits (Region-wise)

```
plt.figure(figsize= (10,10), dpi = 100)

plt.subplot(2,2,1)
plt.title('West')
sns.barplot(x = average(west_df, 'Segment', 'Profit').index,
            y = average(west_df, 'Segment', 'Profit'),
            palette='magma')

plt.subplot(2,2,2)
plt.title('East')
sns.barplot(x = average(east_df, 'Segment', 'Profit').index,
            y = average(east_df, 'Segment', 'Profit'),
            palette='flare')

plt.subplot(2,2,3)
plt.title('Central')
sns.barplot(x = average(central_df, 'Segment', 'Profit').index,
            y = average(central_df, 'Segment', 'Profit'),
            palette='rocket')

plt.subplot(2,2,4)
plt.title('South')
sns.barplot(x = average(south_df, 'Segment', 'Profit').index,
            y = average(south_df, 'Segment', 'Profit'),
            palette='viridis');
```

**However, if we consider average profits, then the story is little different**
* **In West, Corporate has provided the highest average profit**
* **in East and Central, Home Office has provided the highest average profit**
